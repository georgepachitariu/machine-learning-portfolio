### Display augmented images
In this small notebook we visually inspect the effect of augmentation on the images. We want to see that:
1. The same image is both zoomed in more and less and it has the same ratio

```python
import resnet
from resnet import Preprocessing
import numpy as np
import matplotlib.pyplot as plt

def show_batch(ds):
    ds = ds.map(Preprocessing.denormalize)
    
    plt.figure(figsize=(12,10))
    for i, (image, label) in enumerate(ds.take(10)):
        ax = plt.subplot(4,5,i+1)
        plt.imshow(image.numpy().astype(np.uint8))         
        plt.title(label.numpy())
        plt.axis('off')
```

```python
network = resnet.Resnet()
network.load_data(sample_fraction=0.01, only_one=True) # we take only 1 image
network.create_generator(batch_size=1)
show_batch(network.train_augmented_gen)
```

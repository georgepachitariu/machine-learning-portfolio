### Preview augmented images
In this small notebook we visually inspect the effect of augmentation on the images. We want to see that:
1. The code didn't produce any erroneus results: that the image categories are shuffled, the images have the right ratio and are understandable;
2. The augmentation has an effect of enriching the dataset but at the same time it's not too extreme to not make the images not understandable at all;

```python
import alexnet
import numpy as np
import matplotlib.pyplot as plt

def show_batch(ds):
    ds = ds.map(alexnet.Preprocessing.denormalize)
    
    plt.figure(figsize=(12,10))
    for i, (image, label) in enumerate(ds.take(10)):
        ax = plt.subplot(4,5,i+1)
        plt.imshow(image.numpy().astype(np.uint8))         
        plt.title(label.numpy())
        plt.axis('off')
```

```python
network = alexnet.Alexnet()
network.load_data(sample_fraction=0.1)
network.create_generator(batch_size=1)
show_batch(network.train_sample_gen)
```

```python
# we take only 1 to see how the augmentation worked
network = alexnet.Alexnet()
network.load_data(only_one=True)
network.create_generator(batch_size=1)
show_batch(network.train_sample_gen)
```

## Display top-10 predictions before and after training

```python
import numpy as np
import tensorflow as tf
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

import alexnet
import utils
network = alexnet.Alexnet()
network.load_data(sample_fraction=0.01)
network.create_generator(batch_size=32)
```

```python
# 1. Using a non trained model
network.build_model()
utils.display_images_with_predictions(network.validation_gen, network)
```

```python
# 2. Using a trained model
network.load_model(version='v1.4')
utils.display_images_with_predictions(network.validation_gen, network)
```
